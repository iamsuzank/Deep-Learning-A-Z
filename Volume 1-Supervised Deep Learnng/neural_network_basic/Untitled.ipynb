## How Neural Network Learn?
* Suppose we want to predict how much a student will score in Exam based on the features such as 
  * Study hrs
  * Sleep hrs
  * Marks scored in Quiz


```python
from IPython.display import Image,display
display(Image(filename='nn6.JPG'))
```

* When the input features and their corresponding weights are passed to the neuron it makes the prediction as output value(say 83%) but this output value is different than the actual value(93%)
* This difference in output value and actual value is error which is calculated by using the cost function C 
* The value of cost function C must be minimized in order to increase the efficency


```python
from IPython.display import Image,display
display(Image(filename='nn7.JPG'))
```

* In order to minimize the value of cost function C the error is back propagated and the value of weights w1,w2,..wm is tweaked in such a way that minimizes the error or cost function
* Input is multiplied with these updated value of weight and the result is passed forward and output value is calculated 

```python
from IPython.display import Image,display
display(Image(filename='nn8.JPG'))
```

* After updating the value of weights the difference between output value and actual value is very small because the value cost function or error is very low
* In this way we can predict the Percentage of a student in exam
