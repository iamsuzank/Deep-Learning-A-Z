```python
# Dataset Splitting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

## Part 2-Making ANN

```python
# importing keras libraries
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Initializing the ANN
classifier=Sequential()
# adding input layer and first hidden layers with dropout
# classifier.add(Dense(output_dim=6,init='uniform',activation='relu',input_dim=11))
classifier.add(Dense(activation="relu",input_dim=10, units=6, kernel_initializer="uniform"))
classifier.add(Dropout(0.1))
# input_dim is the no of input the first hidden layer takes
# units is  the no of node in the first hidden layer
#general convetntion is that :
#  * no of node in hidden layer=(no of input node+no of output node)/2 i.e. (11+1)/2=6
```

```python
# Adding Second hidden layer
classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))
classifier.add(Dropout(0.1))
```

```python
# Adding output layer
classifier.add(Dense(units=1,activation="sigmoid", kernel_initializer="uniform"))
```

```python
# Compiling the ANN
classifier.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'])
# Optimizer:algorithm to find the optimal weights
# loss:difference between output value and actual value
# metrics:show how weights are updated and accuracy on each epoch
```

## Training ANN

```python
# Fitting the ANN to the training set
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

```python
# Making the prediction and evaluating model
y_pred=classifier.predict(X_test)
# y_pred #gives value in terms of 0 to 1
y_pred=(y_pred>0.5)#if y_pred>0.5 then True else False
# y_pred #gives value in terms of True and False
# making the confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred) 
cm
```

## Evaluating the ANN

```python
# Evaluating the ANN
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense
def build_classifier():
    classifier=Sequential()
    classifier.add(Dense(activation="relu",input_dim=10, units=6, kernel_initializer="uniform"))
    classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))
    classifier.add(Dense(units=1,activation="sigmoid", kernel_initializer="uniform"))
    classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
classifier=KerasClassifier(build_fn=build_classifier,batch_size=10,epochs=100)
accuracies=cross_val_score(estimator=classifier,X=X_train,y=y_train,cv=10,n_jobs=-1)
# cv=10 Fold Cross Validation
# n_jobs=-1:Use all the CPU
mean=accuracies.mean()
variance=accuracies.std()
print('[INFO]:Testing Accuracy={}\n[INFO]:Variance={}'.format(mean,variance))
```

## Tuning the ANN using Grid Search

```python
from sklearn.model_selection import GridSearchCV
def build_classifier(optimizers):
    classifier=Sequential()
    classifier.add(Dense(activation="relu",input_dim=10, units=6, kernel_initializer="uniform"))
    classifier.add(Dense(activation="relu", units=6, kernel_initializer="uniform"))
    classifier.add(Dense(units=1,activation="sigmoid", kernel_initializer="uniform"))
    classifier.compile(optimizer=optimizers,loss='binary_crossentropy',metrics=['accuracy'])
    return classifier
classifier=KerasClassifier(build_fn=build_classifier)
parameters={'batch_size':[25,32],
           'epochs':[100,300],
           'optimizers':['adam','rmsprop']}
grid_search=GridSearchCV(estimator=classifier,
                        param_grid=parameters,
                        scoring='accuracy',
                        cv=10)
grid_search=grid_search.fit(X_train,y_train)
```

```python
best_parameters=grid_search.best_params_
best_accuracy=grid_search.best_score_
print(best_accuracy)
print(best_parameters)
```
